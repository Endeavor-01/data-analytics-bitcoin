```python
splitted = df['Date'].str.split('-', expand=True)

df['year'] = splitted[0].astype('int')
df['month'] = splitted[1].astype('int')
df['day'] = splitted[2].astype('int')

df.head()
```

```python
# Ensure the data types of 'Open', 'High', 'Low', 'Close' are numeric
numeric_columns = ['Open', 'High', 'Low', 'Close']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
data_grouped = df.groupby('year')[numeric_columns].mean()
```

```python
plt.subplots(figsize=(20, 10))
for i, col in enumerate(numeric_columns):
    plt.subplot(2, 2, i+1)
    data_grouped[col].plot.bar()
    plt.title(col)  # Optional: Add a title for each subplot
plt.show()
```

```python
df['is_quarter_end'] = np.where(df['month']%3==0,1,0)
df.head()
```

```python
df['open-close'] = df['Open'] - df['Close']
df['low-high'] = df['Low'] - df['High']
df['target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
```

```python
plt.pie(df['target'].value_counts().values, 
		labels=[0, 1], autopct='%1.1f%%')
plt.show()
```

```python
# Select only numeric columns for the correlation matrix
numeric_df = df.select_dtypes(include='number')
```

```python
# Calculate the correlation matrix
corr_matrix = numeric_df.corr()
```

```python
# Create a boolean mask for correlations greater than 0.9
high_corr = corr_matrix > 0.9
```

```python
plt.figure(figsize=(10, 10))
sb.heatmap(high_corr, annot=True, cbar=False)
plt.show()
```

```python
features = df[['open-close', 'low-high', 'is_quarter_end']]
target = df['target']

scaler = StandardScaler()
features = scaler.fit_transform(features)

X_train, X_valid, Y_train, Y_valid = train_test_split(
	features, target, test_size=0.1, random_state=2022)
print(X_train.shape, X_valid.shape)
```

```python
models = [LogisticRegression(), SVC(kernel='poly', probability=True), XGBClassifier()]

for i in range(3):
   models[i].fit(X_train, Y_train)

   print(f'{models[i]} : ')
   print('Training Accuracy : ', metrics.roc_auc_score(Y_train, models[i].predict_proba(X_train)[:,1]))
   print('Validation Accuracy : ', metrics.roc_auc_score(Y_valid, models[i].predict_proba(X_valid)[:,1]))
   print()
```
